```python
data.rename(columns = {'Revenue (Millions)':'Rev_M'},inplace=True)
```

```python
# os filmes mais bem Avaliados
sorted_df = data.sort_values(by=['Rating'], axis=0, ascending=False)
#mostra as posicoes da tabela
sorted_df[['Title','Director','Rating']].head(10) #5 primeiras posicoes
```

```python
# edgecolor ='black'
#data.hist(column="Year",bins = 30,edgecolor ='black')
# Number of Movies
# Revenue (Millions)

data.Year.plot(kind = 'hist', bins = 40 ,edgecolor ='black', figsize = (9,6))
plt.xlabel("Years")
plt.ylabel("Years")
plt.show()

```

```python
# scatter plot - grafico de dispersao

plt.scatter(data.Year, data.Rating, alpha = 0.28, label = "Movie", color = "red")
plt.xlabel("Years")
plt.ylabel("Ratings")
plt.legend(loc = "lower right")
plt.show()
```

```python
# scatter plot - grafico de dispersao

plt.scatter(data.Year, data.Metascore, alpha = 0.28, label = "Movie", color = "red")
plt.xlabel("Years")
plt.ylabel("Metascore")
plt.legend(loc = "lower right")
plt.show()
```

```python
# scatter plot - grafico de dispersao

plt.scatter(data.Year, data.Rev_M, alpha = 0.28, label = "Movie", color = "red")
plt.xlabel("Years")
plt.ylabel("Revenue (Millions)")
plt.legend(loc = "lower right")
plt.show()
```

```python
# quero ver os filmes que mais faturaram

# ordena DataFrame por Operador de dificuldade convertidos em ordem decrescente
sorted_df = data.sort_values(by=['Rev_M'], axis=0, ascending=False)
#mostra as posicoes da tabela
sorted_df[['Title','Director','Rev_M']].head(10) #5 primeiras posicoes
```

```python
# os piores filmes Avaliados

sorted_df = data.sort_values(by=['Rating'], axis=0, ascending=True)
#mostra as posicoes da tabela
sorted_df[['Title','Director','Year','Rating']].head(10) #5 primeiras posicoes
```
