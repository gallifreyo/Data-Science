```python
import numpy as np
import matplotlib.pyplot as plt

labels = ['13-17', '18-24', '25-34','35-44','45-54','55-64','65+']
size = [10, 14, 19,17,17,13,10]

fig,ax = plt.subplots()

ax.pie(size,labels=labels,autopct='%1.1f%%',
       shadow = True,startangle=180)

ax.axis('equal')

ax.set_title('Usuarios do Facebook')

plt.show()
```
