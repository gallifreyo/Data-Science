```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

x = np.linspace(0, 20, 500)
y = np.sin(x)
ax.plot(x, y, 'bv',linewidth=5, label = 'sine function', alpha = 0.6 )
ax.legend()

plt.savefig('seno.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

x = np.linspace(0, 10, 200)
y = np.sin(x)
z = np.cos(x)
ax.plot(x, y, 'bp',linewidth=5, label = 'sine function', alpha = 0.6 )
ax.plot(x, z, 'g*',linewidth=3, label = 'cos function', alpha = 0.6 )
ax.legend()

plt.savefig('senocos.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

x = np.linspace(0, 20, 500)
y = np.cos(x)
z = np.tan(x)
ax.plot(x, y, 'b-',linewidth=5, label = 'cos function', alpha = 0.6 )
ax.plot(x, z, 'g-',linewidth=3, label = 'tan function', alpha = 0.8 )
ax.legend()

plt.savefig('costan.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
x = np.arange(0,5,0.2)
y = x
ax.plot(x,y,'r--',linewidth = 2,label = '$y = x$', alpha = 0.6)

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

ax.legend(loc='upper center')

ax.set_yticks([-1,0,1])
ax.set_title('Test plot')


plt.show()
```
